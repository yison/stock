```python
import tushare as ts
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import utils
import os
import datetime


stocks = [
    '601006',
    '600919',
    '601328',
    '600027',
    '600115',
    '603260',
    '600023',
    '600521',
    '600332',
    '603260',
    '601333'
    '01333'
]

file_path = "/home/cloud001/ly/mygithub/stock/data/trading/history/day/tracking_trading_day_model002.csv"
df_raw = pd.read_csv(file_path)
date = datetime.date.today().strftime("%Y-%m-%d")
# date = "2018-04-09"
df = df_raw[df_raw.day == date]
# print (df.head())
df_20M = df[df.large_amount_threshold == 20000000]
df_50M = df[df.large_amount_threshold == 50000000]
df_880k = df[df.large_amount_threshold == 880000]


print ("large amount threshold 50M:")
df_50M_filtered = df_50M[['code', 'type','amount']]
_df_50M = df_50M_filtered.groupby(['code', 'type']).agg('sum')

__df_50M = _df_50M.unstack().fillna(0)
##ignore neutral

__df_50M['amount_diff'] = __df_50M[('amount','buy')] - __df_50M[('amount', 'sell')]
print ("ascending:")
print (__df_50M.sort_values(['amount_diff']))
print ("descending:")
print (__df_50M.sort_values(['amount_diff'], ascending=False))
print ("---------------")

print ("large amount threshold 20M:")

df_20M_filtered = df_20M[['code', 'type','amount']]
_df_20M = df_20M_filtered.groupby(['code', 'type']).agg('sum')

__df_20M = _df_20M.unstack().fillna(0)
# print (__df_20M.head())
# __df_20M[('amount', 'buy')]
##ignore neutral

__df_20M['amount_diff'] = __df_20M[('amount','buy')] - __df_20M[('amount', 'sell')]
print ("ascending:")
print (__df_20M.sort_values(['amount_diff']))
print ("descending:")
print (__df_20M.sort_values(['amount_diff'], ascending=False))
print ("---------------")

print ("large amount threshold 880k:")
df_880k_filtered = df_880k[['code', 'type','amount']]
_df_880k = df_880k_filtered.groupby(['code', 'type']).agg('sum')

__df_880k = _df_880k.unstack().fillna(0)
__df_880k['amount_diff'] = __df_880k['amount','buy'] - __df_880k['amount', 'sell']
print ("ascending:")
print (__df_880k.sort_values(['amount_diff']))
print ("descending:")
print (__df_880k.sort_values(['amount_diff'], ascending=False))
print ("---------------")

print ("50M Warning!!!!!")
print (__df_50M[__df_50M.index.isin(stocks)])
print ("20M Warning!!!")
print (__df_20M[__df_20M.index.isin(stocks)])
print ("880k Warning!")
print (__df_880k[__df_880k.index.isin(stocks)])


df[df.code==2151]
```
